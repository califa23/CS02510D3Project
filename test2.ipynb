```python
# Mapping dictionary
car_type_mapping = {
    '1': 'Two Axle Car',
    '2': 'Two Axle Truck',
    '3': 'Three Axle Truck',
    '4': 'Four Axle Truck',
    '5': 'Two Axle Bus',
    '6': 'Three Axle Bus',
    '2P': 'Park Service'
}

# Apply mapping to the "car-type" column
df['car-type'] = df['car-type'].map(car_type_mapping)
```
